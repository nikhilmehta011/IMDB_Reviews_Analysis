# N-Gram Analysis

### Preprocessing

```python
# Converting into lowercase
df['review'] = df['review'].str.lower()
```

```python
from nltk.corpus import stopwords

stop = set(stopwords.words("english"))

def remove_stopwords(text):
    text = [word.lower() for word in text.split() if word.lower() not in stop]

    return " ".join(text)

df['review'] = df['review'].map(lambda x: remove_stopwords(x))
```

```python
# Removing punctuation
def remove_punct(text):
    table = str.maketrans("", "", string.punctuation)
    return text.translate(table)

df['review'] = df['review'].map(lambda x: remove_punct(x))
```

### Creating positive and negative reviews corpus

```python
def create_corpus(df, sentiment):
    
    corpus =[]
    
    for x in df[df['sentiment']==sentiment]['review'].str.split():
        for i in x:
            corpus.append(i)
    return corpus 
```

```python
pos_corpus = create_corpus(df, 'positive')
neg_corpus = create_corpus(df, 'negative')
```

```python
len(pos_corpus)
```

```python
len(neg_corpus)
```

### Frequently Occuring Words

```python
from collections import defaultdict
```

```python
pos_words = defaultdict(int)
for word in pos_corpus:
    pos_words[word] += 1
    
neg_words = defaultdict(int)
for word in neg_corpus:
    neg_words[word] += 1
```

```python
top_pos_words = sorted(pos_words.items(), key=lambda x: x[1], reverse=True)[:10]
top_neg_words = sorted(neg_words.items(), key=lambda x: x[1], reverse=True)[:10]
```

### Bi-Gram Analysis

```python
import seaborn as sns
from nltk import bigrams
```

```python
pos_bi = list(bigrams(pos_corpus))
neg_bi = list(bigrams(neg_corpus))
```

```python
pos_bi_count = defaultdict(int)
for word in pos_bi:
    pos_bi_count[word] += 1 
    
neg_bi_count = defaultdict(int)
for word in neg_bi:
    neg_bi_count[word] += 1   
```

```python
top_pos_bi = pd.DataFrame(sorted(pos_bi_count.items(), key=lambda x: x[1], reverse=True)[:10])
top_neg_bi = pd.DataFrame(sorted(neg_bi_count.items(), key=lambda x: x[1], reverse=True)[:10])
```
